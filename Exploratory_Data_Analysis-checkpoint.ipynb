## Part 1: Analyze User Purchases and Redemptions

```python
user_balance=pd.read_csv(r'Dataset\user_balance_table.csv')
print('Load success!')
```

```python
#The report_date type is int, now we convert it to datetime type and add it to a new column
user_balance['date'] = pd.to_datetime(user_balance.report_date, format="%Y%m%d")
#set up the datatime day, weekday, week, month, year
user_balance['day'] = user_balance.date.dt.day
user_balance['weekday'] = user_balance.date.dt.weekday
user_balance['week'] = user_balance.date.dt.week
user_balance['month'] = user_balance.date.dt.month
user_balance['year'] = user_balance.date.dt.year
user_balance
```

### 1. The Daily Total Amount of Purchases and Redemptions

```python
day_total = user_balance[['date', 'total_purchase_amt', 'total_redeem_amt']].groupby('date').sum()
day_total.reset_index(inplace=True)
day_total
```

```python
fig = plt.figure(figsize=(20, 5))
plt.plot(day_total.date, day_total.total_purchase_amt, label='Total Purchases')
plt.plot(day_total.date, day_total.total_redeem_amt, label='Total Redemptions')

#Now set up title, labels, and legend
plt.title('The Daily Total Amount of Purchases and Redemptions', fontsize = 20)
plt.xlabel('Time', fontsize = 15)
plt.ylabel('Total Amount', fontsize = 15)
plt.legend(loc=1)
plt.show()
```

As you can see, we can divide the total amount of purchases and redemptions into 4 time periods:  
1. From `2013-07` to `2013-10`: The total purchases and redemptions almost unchanged.   
2. From `2013-11` to `2014-01`: A certain increase in total amount of purchases and redemptions.
3. From `2014-02` to `2014-04`: The total purchases and redemptions fluctuate significantly.  
4. From `2014-04` to `2014-09`: The trends of total purchases and redemptions began to stabilize, and began to show more regular changes.

### 2. The Total Amount of Purchases and Redemptions from 2014-04 to 2014-09

Since the data from `2014-04` to `2014-09` is more stable, and shows more regular changes. It requires more in-depth analysis to solving our final goal.

```python
total414 = day_total[day_total.date >= pd.to_datetime('2014-04-01')]
total414
```

```python
total414['day'] = total414.date.dt.day
total414['weekday'] = total414.date.dt.weekday
total414['week'] = total414.date.dt.week
total414['month'] = total414.date.dt.month
total414['year'] = total414.date.dt.year
total414
```

```python
fig = plt.figure(figsize=(20, 5))
plt.plot(total414.date, total414.total_purchase_amt, label='Total Purchases')
plt.plot(total414.date, total414.total_redeem_amt, label='Total Redemptions')

#Now set up title, labels, and legend
plt.title('The Total Amount of Purchases and Redemptions from 2014-04 to 2014-09', fontsize = 20)
plt.xlabel('Time', fontsize = 15)
plt.ylabel('Total Amount', fontsize = 15)
plt.legend(loc=1)
plt.show()
```

### 3. The Total Amount of Purchases and Redemptions in Each Month from 2013-07 to 2014-09

```python
months = ['2013-07-01', '2013-08-01', '2013-09-01', '2013-10-01', '2013-11-01', '2013-12-01', '2014-01-01',  
        '2014-02-01', '2014-03-01', '2014-04-01', '2014-05-01', '2014-06-01', '2014-07-01', '2014-08-01', '2014-09-01']
```

```python
fig = plt.figure(figsize = (15,40))

m = 14
p = 1
num = 1
fig.suptitle("The Total Amount of Purchases and Redemption in Each Month from 2013-07 to 2014-09", fontsize=16)
for i in range(len(months)-1):
    month_amount = day_total[(day_total.date >= pd.to_datetime(months[i])) & (day_total.date < pd.to_datetime(months[i+1]))]
    plt.subplot(m, p, num)
    
    plt.plot(month_amount.date, month_amount.total_purchase_amt)
    plt.plot(month_amount.date, month_amount.total_redeem_amt)
    plt.legend(['Total Purchases','Total Redemptions'], loc = 1)
    plt.xlabel('Time')
    plt.ylabel('Total Amount')
    num+=1

plt.tight_layout()

fig.subplots_adjust(top=0.97)

plt.show()
```

From the line plots of `2` and `3`, it is obvious that the total amount puchases and redemptions in each month have a certain similarity:  
1. The change between the total purchases and redemptions is roughly weekly, with 4 peaks and 4 troughs.
2. At the beginning of the month, the total amount of redemption is often lower than the total amount of purchases, but at the end of the month, vice versa.

As can be seen from the above tables, the distribution of the toal purchases and redemptions within a week is relatively concentrated with a small number of extreme values. And it contain a weekly trend after all. I also found out that:
1. From `Monday` to `Thursday`, the total purchases and redemptions are very close.
2. On `Friday`, the total purchases and redemptions had decreased.
3. On `Saturday` and `Sunday`,  the total purchases and redemptions at the lowest amount.

### 5. The Impact of Date on Total Purchases and Total Redemption

```python
fig = plt.figure(figsize = (14,70))

m = 14
p = 2
num = 1
fig.suptitle("The Average Total Amount of Purchases and Redemption in Each Month from 2013-07 to 2014-09", fontsize=16)

for i in range(len(months)-1):
    month_amount = day_total[(day_total.date >= pd.to_datetime(months[i])) & (day_total.date < pd.to_datetime(months[i+1]))]
    month_amount['day'] = month_amount.date.dt.day
    avg_day = month_amount[['total_purchase_amt', 'total_redeem_amt', 'day']].groupby('day', as_index = False).mean()
    plt.subplot(m, p, num)
    plt.title("The Average Total Amount of Purchases Start at {0}".format(months[i]))
    ax = sns.barplot(x='day', y='total_purchase_amt', data=avg_day)
    ax = sns.lineplot(x='day', y='total_purchase_amt', data=avg_day)
    num+=1
    
    plt.subplot(m, p, num)
    plt.title("The Average Total Amount of Redemptions Start at {0}".format(months[i]))
    ax = sns.barplot(x='day', y='total_redeem_amt', data=avg_day)
    ax = sns.lineplot(x='day', y='total_redeem_amt', data=avg_day)
    num+=1

plt.tight_layout()

fig.subplots_adjust(top=0.97)

plt.show()
```

There are findings from above plots:
1. On the day of holidays, such as Mid-Autumn Festival, National Days, Qingming Festival..., there is some unusual activity in the average of total purchases and redemptions.
2. From `2014-04` to `2014-08`, the average of total purchases and redemptions showed a certain periodic and regular changes. At the beginning of the week, people tended to purchase, and in the middle of the week, they tend to redeem it. But the transaction volume on weekends is generally small.

### 6. Holidays 

The above analysis shows that from `April 2014`, the total purchases and redemptions of Yu'e Bao began to show a relatively regular change, so in consideration of holidays, only the holidays `after April 2014` are considered.

```python
#Qingming Festival
QM = day_total[(day_total.date >= pd.to_datetime('2014-04-05')) & (day_total.date < pd.to_datetime('2014-04-08'))]

#Labour's Day
LD = day_total[(day_total.date >= pd.to_datetime('2014-05-01')) & (day_total.date <  pd.to_datetime('2014-05-04'))]

#Dragon Boat Festival
DB = day_total[(day_total.date >=  pd.to_datetime('2014-05-31')) & (day_total.date <  pd.to_datetime('2014-06-03'))]

#Average
avg4to6 = day_total[(day_total.date >=  pd.to_datetime('2014-04-01')) & (day_total.date <  pd.to_datetime('2014-07-01'))]
```

```python
fig = plt.figure(figsize=(15,5))
x = ['QM', 'LD', 'DB', 'Avg from April to July']
plt.subplot(1,2,1)
y = [np.mean(QM.total_purchase_amt), np.mean(LD.total_purchase_amt), 
     np.mean(DB.total_purchase_amt),np.mean(avg4to6.total_purchase_amt)]
plt.bar(x, y)
plt.title('The Average Total Purhcases of Each Holidays')
plt.ylabel('Total Amount')

plt.subplot(1,2,2)
y = [np.mean(QM.total_redeem_amt), np.mean(LD.total_redeem_amt), 
     np.mean(DB.total_redeem_amt),np.mean(avg4to6.total_redeem_amt)]
plt.bar(x, y)
plt.title('The Average Total Redemptions of Each Holidays')
plt.ylabel('Total Amount')
```

As can be seen from the bar charts, the transaction volume in holidays is signficantly lower than usual.

### 2. Total Purchases and Redemptions by Cities

```python
cities = upf.groupby(['date', 'city'], as_index=False)['total_purchase_amt', 'total_redeem_amt'].sum()
cities
```

```python
fig = plt.figure(figsize=(15,8))
for i in np.unique(cities.city):
    ax = sns.kdeplot(cities[cities.city == i]['total_purchase_amt'], label = i)
plt.legend(loc=0)
plt.title('The Kernel Density Plot of Total Purchases in Different Cities')
plt.xlabel('Time')
plt.ylabel('Total Amount')
plt.show()
```

```python
fig = plt.figure(figsize=(15,8))
for i in np.unique(cities.city):
    ax = sns.kdeplot(cities[cities.city == i]['total_redeem_amt'], label = i)
plt.legend(loc=0)
plt.title('The Kernel Density Plot of Total Redemptions in Different Cities')
plt.xlabel('Time')
plt.ylabel('Total Amount')
plt.show()
```
